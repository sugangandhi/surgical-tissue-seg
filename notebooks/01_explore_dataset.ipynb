# 📊 CholecSeg8k Dataset Exploration
Explore the dataset: class distribution, sample frames, mask visualization.

```python
from datasets import load_dataset
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
ds = load_dataset('minwoosun/CholecSeg8k', trust_remote_code=True)
print(ds)
```

```python
# Visualize a sample
sample = ds['train'][0]
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].imshow(sample['image'])
axes[0].set_title('Original Frame')
axes[1].imshow(sample['mask'])
axes[1].set_title('Tissue Mask')
plt.tight_layout()
plt.savefig('../results/sample_frame.png', dpi=150)
plt.show()
```

```python
# Class distribution
CLASS_NAMES = ['background','abdominal_wall','liver','gi_tract','fat',
               'grasper','connective_tissue','blood','cystic_duct',
               'l_hook','gallbladder','hepatoduodenal_ligament','clipper']

class_pixel_counts = np.zeros(13)
for sample in ds['train'].select(range(200)):  # Sample 200 frames
    mask = np.array(sample['mask'])
    for i in range(13):
        class_pixel_counts[i] += (mask == i).sum()

plt.figure(figsize=(12, 5))
plt.bar(CLASS_NAMES, class_pixel_counts / class_pixel_counts.sum() * 100)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Pixel %')
plt.title('Class Distribution in CholecSeg8k (200 frames)')
plt.tight_layout()
plt.savefig('../results/class_distribution.png', dpi=150)
plt.show()
```
